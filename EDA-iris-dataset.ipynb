### Importing other Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Dataset

```python
df = sns.load_dataset('iris')
df.head()
```

### Univariate Analysis

```python
# Creating separate data files for setosa, virginica, versicolor 

df_setosa = df.loc[df['species']=='setosa']
df_virginica = df.loc[df['species']=='virginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Sepal length')
plt.show()
```

### Observation: 
- We can observe that majority of setosa species have sepal length under 5.0 however there is a datapoint that lies between 5.0 and 5.5 which can be treated as an outlier.
- Sepal length of virginica ends at 7.0 and versicolor species starts at 7.0 , it can help us to create a separation between two species.

#### Bivariate Analysis

```python
sns.FacetGrid(df,hue='species',height=3).map(plt.scatter,"sepal_length","sepal_width").add_legend();
plt.show()
```

### Observation
- There is a separation between setosa species datapoints and other two species which can be helpful.
- There are mixed datapoints having sepal length between 5.5-7 and sepal width between 2.5-3
- There are some outliers present beyond sepal length 8
